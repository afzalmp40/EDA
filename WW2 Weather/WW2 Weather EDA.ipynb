```python
# importing data
df=pd.read_csv('WW2 Weather.csv', parse_dates=['Date'])
df.head()
```

```python
df.info()
```

### ***There are a lot of features with more than 50% missing values***

### Removing features with high missing values

```python
# getting null value count in each feature
null_count=df.isnull().sum()

# converting to percentage of values missing
null_count=null_count*100/df.shape[0]

# getting list of useless columns
useless_columns= list( null_count[null_count>50].index )

useless_columns
```

```python
# dropping useless columns
df.drop(columns=useless_columns, axis=1, inplace=True)
```

```python
# checking again for missing values
df.info()
```

```python
df.isnull().sum()
```

    There are features with some missing values.
    We can try to impute those values.

### Data dictionary

```python
pd.read_csv('data_dict.csv', index_col='Feature')
```

### ***Looks like 5 features are repeating (represented with different units)*** 
### ***Lets drop these repeating features that also have a few missing values***

```python
df.drop(columns=['PRCP','MAX','MIN','MEA','SNF'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# checking again for missing values
df.info()
```

    Looks like Precip and Snowfall are numerical columns but stored as object 
    So lets convert them to float64

#### **df[['Precip', 'Snowfall']].astype(np.float64)**
    Running the above command gives an error

#### **ValueError: could not convert string to float: 'T'**
    
    Seems like there is a string value in the feature

### Handling snowfall feature

```python
df.Snowfall.value_counts()
```

    We found that '#VALUE!' was a value in Snowfall that was non numerical and hence whole Snowfall feature became object

```python
# replacing '#VALUE!' with np.nan for easier processing
df.Snowfall=df.Snowfall.replace('#VALUE!', np.nan).astype('float64')
```

### handling Precip feature

```python
df.Precip.value_counts()
```

    We found that 'T' was a value in Precip that was non numerical and hence whole Precip feature became object

```python
# replacing 'T' with np.nan
df.Precip=df.Precip.replace('T', np.nan).astype('float64')
```

### Now our data has been converted to numerical

```python
df.info()
```

### Lets have a look at the distributions

```python
fig, ax=plt.subplots(figsize=(18,10), dpi=150)
df.hist(ax=ax, bins=63, alpha=0.7, edgecolor='black')
plt.show()
```

### **Imputing missing values**
    Distribution for Precip and Snowfall is heavily right skewed.
    Lets impute the missing values in each feature with its corresponding median

```python
columns_to_impute=['Snowfall', 'Precip']

for col in columns_to_impute:
    df.replace( to_replace={ col: { np.nan: df[col].median() } }, inplace=True )
```

```python
df.info()
```

    Looks like all features are discrete

    It appears Precipitation and Snowfall were recorded in inches and then converted to mm
    Temperatures were recorded in farenheit and converted to celsius
    So lets convert these values to original metrics for easier understanding 

```python
for col in ['Precip','Snowfall']:
    # conversion from mm to inches
    df[col]= ( df[col]/25.4 ).astype(np.float64)
```

```python
for col in ['MaxTemp','target', 'MeanTemp']:
    # conversion from celsius to farenheit
    df[col]= ( df[col] * 9/5 +32  ).astype(np.float64)
```

    And temperatures are in Farenheit

### Lets see temperature variations throughout the timeframe

```python
fig,ax=plt.subplots(figsize=(25,5), dpi=100)

col_dict={'red':'MaxTemp','orange':'MeanTemp',  'yellow':'target'}

for color,col in col_dict.items():
    
    ax=sns.lineplot(
        data=df,
        x='Date',
        y=col,
        color=color, label=col
    )
    
ax.set_xlabel('Time period( month-year )')
ax.set_ylabel('Temperatures')
ax.set_title('Temperature variations throughout the timeframe')
plt.xticks(rotation=90)
ax.legend(loc='best')
plt.show()
```

    Looks like 1942 had lowest temperatures as compared to others

### **Plotting data against the Date feature is taking a lot of time and hence i grouped the data and analysed it**

## Lets group the dataset based on year and month and aggregate using mean()

```python
df_YM_grouped=df.groupby(['YR','MO']).mean().iloc[:,1:6].reset_index()
df_YM_grouped['mo_yr']=df_YM_grouped.MO.astype(str) + '-' + df_YM_grouped.YR.astype(str)
df_YM_grouped
```

### Now lets see the temperature variations for the monthly grouped period (1940 till 1945)

```python
fig,ax=plt.subplots(figsize=(15,5), dpi=150)

col_dict={'red':'MaxTemp','orange':'MeanTemp',  'yellow':'target'}

for color,col in col_dict.items():
    
    ax=sns.lineplot(
        data=df_YM_grouped,
        x='mo_yr',
        y=col,
        color=color, label=col
    )
    
ax.set_xlabel('Time period( month-year )')
ax.set_ylabel('Temperatures')
ax.set_title('Temperature variations throughout the timeframe')
plt.xticks(rotation=90)
ax.legend(loc='best')
plt.show()
```

### ***Looks like 1942 had lowest temperature drop till date***
### *Also looks like temperatures generally drop after september*

### Lets see Precipitation variations for the monthly grouped period (1940 till 1945)

```python
fig,ax=plt.subplots(figsize=(16,3), dpi=200)


for color,col in col_dict.items():
    
    ax=sns.lineplot(
        data=df_YM_grouped,
        x='mo_yr',
        y='Precip',
        color='blue'
    )
    
ax.set_xlabel('Time period( month-year )')
ax.set_ylabel('Precipitation')
ax.set_title('Precipitation variations throughout the timeframe')
plt.xticks(rotation=90)
plt.show()
```

### Lets see Snowfall variations for the monthly grouped period (1940 till 1945)

```python
fig,ax=plt.subplots(figsize=(16,3), dpi=200)


for color,col in col_dict.items():
    
    ax=sns.lineplot(
        data=df_YM_grouped,
        x='mo_yr',
        y='Snowfall',
        color='cyan'
    )
    
ax.set_xlabel('Time period( year )')
ax.set_ylabel('Snowfall')
ax.set_title('Snowfall variations throughout the timeframe')
plt.xticks(rotation=90)
plt.show()
```
